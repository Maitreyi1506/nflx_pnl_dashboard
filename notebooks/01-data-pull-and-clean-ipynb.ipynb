```python
import yfinance as yf
import pandas as pd

ticker  = "NFLX"
data = yf.download(ticker, start="2015-01-01", progress=False)
data.columns = data.columns.get_level_values(0)
print(data.columns)

prices = data[['Close']].rename(columns={'Close': 'Price'})
prices['Return'] = prices['Price'].pct_change()
prices.columns.name = None
print(prices.tail())

# prices.to_csv("nflx_prices.csv")
```
